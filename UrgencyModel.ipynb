```python
#Import Libraries
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizer, BertForSequenceClassification, get_scheduler
from torch.optim import AdamW
from sklearn.model_selection import train_test_split
import numpy as np
import re

# Download Sentiment140 Dataset (Training set only)
!wget http://cs.stanford.edu/people/alecmgo/trainingandtestdata.zip
!unzip trainingandtestdata.zip

# Load and Prepare the Data
sentiment140_path = '/content/training.1600000.processed.noemoticon.csv'
df = pd.read_csv(sentiment140_path, encoding='latin-1', header=None)
```

```python
# Dataset columns:
# 0: Sentiment (0 = negative, 2 = neutral, 4 = positive)
# 5: Tweet text
df = df[[0, 5]]
df.columns = ['sentiment', 'text']

# Map Sentiment Values
sentiment_mapping = {0: 0, 2: 1, 4: 2}
df['sentiment'] = df['sentiment'].map(sentiment_mapping)

# Clean the text
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)  # Remove URLs
    text = re.sub(r"[^a-zA-Z\s]", "", text)  # Remove punctuation
    text = re.sub(r"\s+", " ", text)  # Remove extra spaces
    return text.strip()

df['text'] = df['text'].apply(clean_text)

# (Optional) Reduce size if needed (training on 200k for faster training)
df = df.sample(n=200000, random_state=42)

# 5. Tokenizer
model_name = "bert-base-multilingual-cased"
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python
# Dataset Class
class SentimentDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]

        encoding = self.tokenizer(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            padding="max_length",
            truncation=True,
            return_tensors="pt",
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# 7. Split Train and Validation Sets
X_train, X_val, y_train, y_val = train_test_split(df['text'], df['sentiment'], test_size=0.1, random_state=42)

train_dataset = SentimentDataset(X_train.tolist(), y_train.tolist(), tokenizer, max_len=64)
val_dataset = SentimentDataset(X_val.tolist(), y_val.tolist(), tokenizer, max_len=64)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)
```

```python
# 8. Load Pretrained Multilingual BERT
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=3)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# 9. Optimizer and Scheduler
optimizer = AdamW(model.parameters(), lr=2e-5)

num_epochs = 3
num_training_steps = num_epochs * len(train_loader)
lr_scheduler = get_scheduler(
    "linear",
    optimizer=optimizer,
    num_warmup_steps=0,
    num_training_steps=num_training_steps,
)

# 10. Training Loop
from tqdm.auto import tqdm

progress_bar = tqdm(range(num_training_steps))

model.train()
for epoch in range(num_epochs):
    for batch in train_loader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)

# 11. Save Trained Model and Tokenizer
model_save_path = "/content/bert_sentiment140_finetuned"
model.save_pretrained(model_save_path)
tokenizer.save_pretrained(model_save_path)

print("Model and tokenizer saved successfully at:", model_save_path)
```

```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader, random_split
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from sklearn.preprocessing import LabelEncoder
import numpy as np

# 1. Load Dataset
df = pd.read_csv('realistic_feedback.csv')

# 2. Label Encoding
sentiment_encoder = LabelEncoder()
df['sentiment_label'] = sentiment_encoder.fit_transform(df['sentiment'])

# 3. Tokenization
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

class FeedbackDataset(Dataset):
    def __init__(self, texts, sentiments, urgencies):
        self.texts = texts
        self.sentiments = sentiments
        self.urgencies = urgencies

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = tokenizer(
            self.texts[idx],
            truncation=True,
            padding='max_length',
            max_length=128,
            return_tensors='pt'
        )
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(self.urgencies[idx], dtype=torch.long)
        }

# 4. Prepare Dataset
dataset = FeedbackDataset(
    texts=df['feedback_text'].tolist(),
    sentiments=df['sentiment_label'].tolist(),
    urgencies=df['urgent'].tolist()
)

# 5. Train/Validation Split
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

# 6. Load Pretrained BERT Model for Binary Classification
model = BertForSequenceClassification.from_pretrained('bert-base-multilingual-cased', num_labels=2)

# 7. Training Arguments
training_args = TrainingArguments(
    output_dir='./results',
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    logging_dir='./logs'
)

# 8. Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset
)


# 9. Train
trainer.train()

# 10. Save Model
model.save_pretrained('/content/bert-urgency-classifier')
tokenizer.save_pretrained('/content/bert-urgency-classifier')

print(" Model fine-tuned and saved!")
```

```python
from sklearn.metrics import classification_report
import numpy as np

# 1. Run predictions on the validation set
outputs = trainer.predict(val_dataset)
preds = np.argmax(outputs.predictions, axis=1)
labels = outputs.label_ids

# 2. Print classification report
print(classification_report(labels, preds, target_names=["Not Urgent", "Urgent"]))
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification

# Load model + tokenizer
model_path = '/content/bert-urgency-classifier'  # Change if different
tokenizer = BertTokenizer.from_pretrained(model_path)
model = BertForSequenceClassification.from_pretrained(model_path)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

# Inference function
def predict_urgency(text):
    encoded = tokenizer(
        text,
        return_tensors='pt',
        truncation=True,
        padding='max_length',
        max_length=128
    ).to(device)

    with torch.no_grad():
        outputs = model(**encoded)
        probs = torch.nn.functional.softmax(outputs.logits, dim=1)
        predicted_class = torch.argmax(probs, dim=1).item()
        confidence = probs[0][1].item()  # Confidence for "Urgent" class

    return {
        "text": text,
        "is_urgent": bool(predicted_class),
        "confidence": round(confidence, 3),
        "label": "Urgent" if predicted_class == 1 else "Not Urgent"
    }
```

```python
# Example usage
if __name__ == "__main__":
    feedback = "الطلب اتأخر جدًا ومحدش رد عليا"
    result = predict_urgency(feedback)
    print(result)
```

```python
print(predict_urgency("الطلب اتأخر ومحدش رد عليا"))       # Expected: Urgent
print(predict_urgency("شكرا جدًا، التطبيق ممتاز"))         # Expected: Not Urgent
```
